```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = 'https://www.imsdb.com/scripts/Avengers,-The-(2012).html'
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
results = soup.find_all('b')
```

```python
ls = list()
```

```python
for result in results:
    name = result.text.strip()
    if not name.startswith(')') and len(name.split(' '))<4:
        ls.append(name.split(':')[0])
```

```python
len(ls)
```

```python
ls
    
```

```python
df = pd.DataFrame(ls, columns=['Lines'])
```

```python
df.to_csv('The_Avengers.csv', index=False)
```
